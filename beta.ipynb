```python
import pandas as pd
from hisdata import Stock, Index
from indexconstituents import IndexConstituents as ic
```

```python
from_period = 1000
```

```python
bm = Index('bse_sensex').close[-from_period:]
```

```python
def get_beta(stk):
    stk_pct_change = stk.pct_change()
    bm_pct_change = bm.pct_change()
    return bm_pct_change.cov(stk_pct_change)/bm_pct_change.var()
```

```python
stks = ic().bse_200
```

```python
#betas = {'bsecode': bsecode, 'company_name': company_name, 'beta': get_beta(close)
#          for bsecode, company_name, close in zip(stks.bsecode, stks.ix[bsecode, 'company_name'],
#                                                  Stock(

betas = {}
betas['bsecode'] = []
betas['company_name'] = []
betas['beta'] = []
for bsecode in stks.bsecode:
    s = Stock(bsecode)
    betas['bsecode'].append(bsecode)
    betas['company_name'].append(stks.ix[bsecode, 'company_name'])
    betas['beta'].append(get_beta(s.close[-from_period:]))
```

```python
betas
```

```python
betas_df = pd.DataFrame.from_dict(betas)
betas_df = betas_df.set_index('bsecode')
```
