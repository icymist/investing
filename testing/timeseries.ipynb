```python
from pandas import *
from numpy.random import randn
set_option('display.max_rows', 100)
```

```python
rng = date_range('2013-01-01', periods=60, freq='B')
```

```python
df = DataFrame(range(1,61), index=rng)
```

```python
df
```

```python
ra = rolling_mean(df, 2)
```

```python
resampled = ra.asfreq('D', method='pad')
resampled
```

```python
pc = resampled.pct_change()
pc
```

```python
pc.asfreq('B').resample('MS', how='mean')
```

```python
ra2 = rolling_mean(df, 2)
pc2 = ra2.pct_change()
```

```python
pc2.resample('MS', how='mean')
```
